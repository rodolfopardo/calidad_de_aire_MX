```python
#Para cualquier persona, la base es bastante poco entendible
#Vamos a cambiar la nomenclatura de las columnas

columnas = ["Fecha", "Hora","Aguilas", "Atemajac", "Centro", "Loma Dorada", "Miravalle", "Oblatos", "Tlaquepaque", "Vallarta"]

air_quality.columns

air_quality.columns = columnas

air_quality.head(3)
```

```python
#Comenzamos a limpiar la base y sus valores NaN 

air_quality.fillna(0)
```

```python
#Vamos a convertir ahora los tipos de variables 

air_quality.dtypes
```

```python
air_quality['Fecha'] = pd.to_datetime(air_quality['Fecha'], format='%Y-%m-%d')
air_quality.dtypes
```

```python
#En este ejercicio, vamos a guardar en 3 csv diferentes las tres centrales con el mayor promedio de contaminación tomada

aguilas = air_quality.filter(['Aguilas'])
centro = air_quality.filter(['Centro'])
loma_dorada = air_quality.filter(['Loma Dorada'])

#Cada una a un csv aparte 

aguilas.to_csv('registro_aguilas.csv', index = False)
centro.to_csv('registro_centro.csv', index = False)
loma_dorada.to_csv('registro_lomadorada', index= False)
```
